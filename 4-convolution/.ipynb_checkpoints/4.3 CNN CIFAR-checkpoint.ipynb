## 1. Loading and normalizing CIFAR10

Using ``torchvision``, it’s extremely easy to load CIFAR10.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import torch
import torchvision
import torchvision.transforms as transforms
```

The output of torchvision datasets are PILImage images of range [0, 1].
We transform them to Tensors of normalized range [-1, 1]

**NB** Modify the code below to only use a small part of the dataset if your computer is very slow.

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5),
                          (0.5, 0.5, 0.5))]
)

# Load dataset
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)

classes = ('plane', 'car', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck')

used_categories = range(len(classes))

## USE CODE BELOW IF YOUR COMPUTER IS TOO SLOW
reduce_dataset = True
if reduce_dataset:
    used_categories = (3, 5) # cats and dogs

    classes = [classes[i] for i in used_categories]
    new_train_data = []
    new_train_labels = []

    new_test_data = []
    new_test_labels = []
    for i, t in enumerate(used_categories):
        new_train_data.append(trainset.data[np.where(np.array(trainset.targets) == t)])
        new_train_labels += [i for _ in range(new_train_data[-1].shape[0])]

        new_test_data.append(testset.data[np.where(np.array(testset.targets) == t)])
        new_test_labels += [i for _ in range(new_test_data[-1].shape[0])]

    new_train_data = np.concatenate(new_train_data, 0)
    trainset.train_data = new_train_data
    trainset.train_labels = new_train_labels

    new_test_data = np.concatenate(new_test_data, 0)
    testset.test_data = new_test_data
    testset.test_labels = new_test_labels
```

```python
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4,
                                          shuffle=True, num_workers=2)
testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                         shuffle=True, num_workers=2)
train_data_iter = iter(trainloader)
test_data_iter = iter(testloader)
print('used classes:', classes)
```

* ### Functions for training on the GPU
* * Moves variables to GPU memory.

```python
def get_variable(x):
    """ Converts tensors to cuda, if available. """
    if torch.cuda.is_available():
        return x.cuda()
    return x

def get_numpy(x):
    """ Get numpy array for both cuda and not. """
    if torch.cuda.is_available():
        return x.cpu().data.numpy()
    return x.data.numpy()
```

### Function for calculating dimensions
* Important for managing dimensions between convolution and pooling layers

```python
def compute_conv_dim(dim_size, kernel_size=5, stride=1, padding=0):
    return int((dim_size - kernel_size + 2 * padding) / stride + 1)

def compute_pool_dim(dim_size, kernel_size=2, stride=2):
    return int((dim_size - kernel_size) / stride + 1)
```

### Set hyper parameters and dimensions
* Required for defining the network architecture

```python
# Conv layers hyper params
channels = trainset.data.shape[3]
kernel_size = 5
conv1_channels_out = 6
conv2_channels_out = 16
pool_kernel_size = 2
pool_stride = 2

# First conv layer + pooling
dim_size = trainset.data.shape[1]
conv1_dim_out = compute_conv_dim(dim_size)
pool1_dim_out = compute_pool_dim(conv1_dim_out)

# Second conv layer + pooling
conv2_dim_out = compute_conv_dim(pool1_dim_out)
pool2_dim_out = compute_pool_dim(conv2_dim_out)

# Calculate dimension going into dense layers
conv_total_size = conv2_channels_out * pool2_dim_out * pool2_dim_out

# Fully connected layer dimensions
dense1_dim_out = 120
dense2_dim_out = 84
num_classes = len(classes)

print("Conv+Pool (1) out_size:", conv1_channels_out * pool1_dim_out * pool1_dim_out)
print("Conv+Pool (2) out_size:", conv_total_size)
print("-"*30)
print("Dense (1) out_size:", dense1_dim_out)
print("Dense (2) out_size:", dense2_dim_out)
print("Dense (3) out_size:", num_classes)
```

```python
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
from torch.nn import Linear, Conv2d, BatchNorm2d, MaxPool2d, Dropout2d
from torch.nn.functional import relu, elu, relu6, sigmoid, tanh, softmax

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # Convolution and pooling layers
        self.conv1 = nn.Conv2d(channels, conv1_channels_out, kernel_size)
        self.conv2 = nn.Conv2d(conv1_channels_out, conv2_channels_out, kernel_size)
        self.pool = nn.MaxPool2d(pool_kernel_size, pool_stride)
        
        # Dense layers
        self.dense1 = nn.Linear(conv_total_size, dense1_dim_out)
        self.dense2 = nn.Linear(dense1_dim_out, dense2_dim_out)
        self.dense3 = nn.Linear(dense2_dim_out, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, conv_total_size)
        x = F.relu(self.dense1(x))
        x = F.relu(self.dense2(x))
        x = self.dense3(x)
        return x
    
net = Net()

if torch.cuda.is_available():
    print('##converting network to cuda-enabled')
    net.cuda()
print(net)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()  # Your code here!
optimizer = optim.Adam(net.parameters(), lr=0.001)  # Your code here!
```

## 4. Train the network

**Assignment 3:** Finish the training loop below. 
Start by using a small number of epochs (e.g. 3).
Even with a low number of epochs you should be able to see results that are better than chance.
When everything is working increase the number of epochs to find out how good your network really is.

```python
num_epoch = 3  # Your code here!

for epoch in range(num_epoch):  # loop over the dataset multiple times
    
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs
        inputs, labels = data
        
        # wrap them in Variable
        inputs, labels = get_variable(inputs), get_variable(labels)

        # zero the parameter gradients
        output = net(inputs)
        loss = criterion(output, labels)
        optimizer.zero_grad()
        
        # forward + backward + optimize
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 1000 == 999:    # print every 1000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 1000))
            running_loss = 0.0

print('Finished Training')
```

* Construct an example showing explicitly that in a network with a sigmoid output layer, the output activations  aLj  won’t always sum to 1.

**Answer:**

We consider a neural net with only one class, i.e. a single neuron in the output layer, with a sigmoid activation function.

By the definition of the sigmoid function, the output is always such that $$0 < \sigma(z) < 1$$, and as such the sum over all outputs, i.e. the sum over the single output neuron, is therefore less than 1.
